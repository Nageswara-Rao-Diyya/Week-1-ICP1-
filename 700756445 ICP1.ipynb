```python
def fullname(first_name, last_name):
    return first_name + " " + last_name

def string_alternative(input_string):
    return input_string[::2]

def main():

    first_name = input("Enter your first name: ")
    last_name = input("Enter your last name: ")

    full_name = fullname(first_name, last_name)
    print(f"Full Name: {full_name}")

    alt_string = string_alternative(full_name)
    print(f"String Alternative: {alt_string}")

if __name__ == "__main__":
    main()
```

```python

def calculate_word_count(input_file, output_file):
    word_count = {}

    try:
        with open(input_file, 'r') as infile:
            lines = infile.readlines()
    except FileNotFoundError:
        print(f"File '{input_file}' not found. Creating an empty file.")

        open(input_file, 'w').close()
        lines = []

    with open(output_file, 'w') as outfile:

        for line in lines:
            outfile.write(line)
            words = line.split()
            for word in words:
                word = word.strip().strip('.,!?"').capitalize()  # Normalize word
                word_count[word] = word_count.get(word, 0) + 1

        outfile.write("\nWord_Count:\n")
        for word, count in word_count.items():
            outfile.write(f"{word}: {count}\n")

input_file = "input.txt"
output_file = "output.txt"

calculate_word_count(input_file, output_file)

print(f"Word count has been calculated and written to {output_file}.")
```

```python


def inches_to_cm(height):
    return round(height * 2.54, 2)

def convert_heights_nested_loop(heights):
    cm_heights = []
    for height in heights:
        cm_heights.append(inches_to_cm(height))
    return cm_heights

def convert_heights_list_comprehension(heights):
    return [inches_to_cm(height) for height in heights]

heights_in_inches = [150, 155, 145, 148]
heights_in_cm_nested = convert_heights_nested_loop(heights_in_inches)
print("Converted heights using nested loop:", heights_in_cm_nested)
heights_in_cm_comprehension = convert_heights_list_comprehension(heights_in_inches)
print("Converted heights using list comprehension:", heights_in_cm_comprehension)
```
